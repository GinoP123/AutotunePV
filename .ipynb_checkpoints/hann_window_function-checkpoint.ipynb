```python
sample_rate = yin.sample_rate
```

```python
def get_hann_window(N):
    window = np.arange(N) * (2 * np.pi / N)
    return 0.5 * (1 - np.cos(window))
```

```python
plt.plot(get_hann_window(10))
```

```python
W = 1000
H = W // 4
hann_window = get_hann_window(W)
```

```python
sample = yin.get_frequency(500, length=5)
```

```python
len(sample)
```

```python
output = np.zeros_like(sample)
for start_index in range(0, len(sample), H):
    splice = np.ones(W)
    output[start_index:start_index+len(splice)] += (hann_window[:len(splice)] * splice)[:len(output[start_index:start_index+len(splice)])]
```

```python
Audio(output, rate=sample_rate)
```
